# câu 14

```python
df['hour'] = pd.to_datetime(df['HOURS']).dt.hour
df['minute'] = pd.to_datetime(df['HOURS']).dt.minute

df
```

```python
import numpy as np
```

```python
sang = (df['hour'] >= 0) &  (df['hour'] < 12)
chieu = (df['hour'] >= 12) &  (df['hour'] < 18)
toi = (df['hour'] >= 18) &  (df['hour'] < 24)

conditions = [sang,chieu,toi]
timing = ['Sáng','Chiều','Tối']

df['time'] = np.select(conditions, timing)
df
```

```python
df.groupby(['MONTH','time'])['TRANS'].count()
```
